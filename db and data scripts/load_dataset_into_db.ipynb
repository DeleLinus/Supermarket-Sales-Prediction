```python
# import necessary packages
import pandas as pd
# import mysql.connector
from sqlalchemy import create_engine
```

```python
file_path='train.csv'
df=pd.read_csv(file_path)
df.head()
```

```python
path='test.csv'
df2=pd.read_csv(path)
df2.head()
```

```python
path='sample_submission.csv'
df3=pd.read_csv(path)
df3.head()
```

```python
host="localhost"
user="root"
password="delelinus"
database="chukwudi_supermarket"
port = 3306
```

```python
# using sqlalchemy
engine = create_engine(f'mysql+mysqlconnector://{user}:{password}@{host}:{port}/{database}', echo=False)
df.to_sql(name="train", con=engine, index=False, if_exists='replace')
df2.to_sql(name="test", con=engine, index=False, if_exists='replace')
df3.to_sql(name="sample_submission", con=engine, index=False, if_exists='replace')
```
